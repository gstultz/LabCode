```python
from LittleMachine_TDS_Lib import TDS
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
tds_dict = {0.70: ('oct18_17.d02', 'oct18_17.d03', 15269), 
            0.55: ('oct19_17.d01', 'oct19_17.d02', 15801), 
            0.35: ('oct23_17.d01', 'oct23_17.d02', 14886)}
           
```

```python
# tds_filename = 'sep13_17.d01
background_filename = 'sep15_17.d02'
background_sensitivity = 17017
start_temp = 95
total_npts = 820
```

```python
background = TDS(background_filename)
background.load_data(start_temp=start_temp, total_npts=total_npts, background_npts=10)
```

```python
plt.figure(figsize=(16, 9))
ax1 = plt.subplot(121)
ax2 = plt.subplot(122)
for key, item in tds_dict.items():    
    tds = TDS(item[0])
    tds.load_data(start_temp=start_temp, total_npts=total_npts, background_npts=10, show_plot=True)
    tds.background_subtraction(background=background, sensitivity=item[2]/background_sensitivity, show_plot=True)
    ax1.plot(tds.temps, tds.counts, label=str(key) + ' ML Au')
    
    tds = TDS(item[1])
    tds.load_data(start_temp=start_temp, total_npts=total_npts, background_npts=10, show_plot=True)
    tds.background_subtraction(background=background, sensitivity=item[2]/background_sensitivity, show_plot=True)
    ax2.plot(tds.temps, tds.counts, label=str(key) + ' ML Au')

ax1.legend()

ax1.grid()

ax2.legend()

ax2.grid()
```
