```python
from CeyerLibrary import loadAuger, generateDict, index_of
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline

monthDict = generateDict()  #Set up the dictionary
```

```python
# Input file name and location
mainpath = '/Users/qingliu/Dropbox (MIT)/littlemachine/'
filename = 'feb23_98.csv'
```

```python
# Load Data
filepath = mainpath + '19' + filename[6:8] + '/' + monthDict[filename[:3]] + '_' + filename[:3] + '/' + filename
df = pd.read_csv(filepath, comment='X')
```

```python
# Process Data
scan = 65
file = open(filepath + '.a','w')
file.write('IGOR\nWaves Voltage' + chr(scan) + ', Auger' + chr(scan) +'\nBegin\n')
for c,v in zip(df.Voltage.values, df.Auger.values):
    file.write('{}\t{}\n'.format(c, v))
file.write('End\n')
file.close()
print(filename, 'Success!')
```
