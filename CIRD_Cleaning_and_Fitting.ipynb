```python
c.load_data()
```

```python
vector = c.counts
```

```python
print('Detect peaks with minimum height and distance filters.')
indexes = detect_peaks(vector, mph=-1000, mpd=1000, threshold=100, edge=None,
                 kpsh=False, valley=False, show=True, ax=None)
print('Peaks are: %s' % (indexes))
```

```python
counts_sub = vector
```

```python
import numpy as np
highest_index = np.argmax(counts_sub[:])
highest_index
```

```python
averages_highest = []
range_candidate = range((highest_index-20),(highest_index + 21))
```

```python

for n in range_candidate:
        average_highest = np.mean(counts_sub[(n-5):(n+6)])
        averages_highest.append(average_highest)
```

```python
averages_highest
```

```python
averages_highest = np.array(averages_highest)
index_average_beamstart = np.argmax(averages_highest)
index_average_beamstart
```
